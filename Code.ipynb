# 3. FCIT

```python
import pandas as pd
import numpy as np
from fcit import fcit


def nonparametric_fcit_test(X, Y, Z, data):
    """
    X and Y are names of variables.
    Z is a list of names of variables.
    data is a pandas data frame.

    Return a float corresponding to the p-value computed from FCIT.
    """

    # implement code here
    x = data[X].values.reshape(-1,1)
    y = data[Y].values.reshape(-1,1)
    z = data[Z].values
    pval = fcit.test(x, y, z)
    return pval

def main():
    """
    Do not edit this function. This function is used for grading purposes only.
    """

    np.random.seed(0)
    data = newest_data

    print(nonparametric_fcit_test("BI1", "Use1", ["Exp1"], data))
    print(nonparametric_fcit_test("Exp1", "Use1", ["BI1"], data))


if __name__ == "__main__":
    main()
```

```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
from scipy.special import expit

import math

def odds_ratio(X, Y, Z, data):
    data['constant'] = 1
    y_train = data[Y]
    if Z == []:
        X_train = data[['constant',X]]
    else:
        X_train = data[['constant',X,Z[0]]]
    OR = math.exp(sm.Logit(np.asarray(y_train), np.asarray(X_train)).fit(disp=0).params[1])    
    return OR

def compute_confidence_intervals(X, Y, Z, data, num_bootstraps=200, alpha=0.05):    
    Ql = alpha/2
    Qu = 1 - alpha/2
    estimates = []
    
    for i in range(num_bootstraps):
        temp_df_ind = np.random.choice(len(data),replace = True, size = len(data))
        temp_df = data.iloc[temp_df_ind]
        estimates.append(odds_ratio(X,Y,Z,temp_df))
    
    sorted_estimates = sorted(estimates)
    q_low = np.percentile(sorted_estimates,Ql*100)
    q_up = np.percentile(sorted_estimates,Qu*100)
    
    return q_low, q_up
```

```python
copy1 = newest_data.copy()
copy1['Inc1'] = copy1['Inc1'].apply(lambda x: 0 if x == 1 or x == 2 or x==3  else 1)
copy1['Exp1'] = copy1['Exp1'].apply(lambda x: 0 if x == 1 or x == 2 or x==3  else 1)
copy1['Use1'] = copy1['Use1'].apply(lambda x: 0 if x == 1 or x == 2 or x==3  else 1)
copy1
np.random.seed(200)
print(odds_ratio("BI1", "Use1", ["Exp1"], copy1),compute_confidence_intervals("BI1", "Use1", ["Exp1"], copy1))
print(odds_ratio("Exp1", "Use1", ["BI1"], copy1), compute_confidence_intervals("Exp1", "Use1", ["BI1"], copy1))
```

# 5. Backdoor adjustment

**5.1 Load the cleaned data here**

```python
newest_data = new_data.dropna().reset_index(drop=True)
newest_data = newest_data.apply(pd.to_numeric)
newest_data['intercept'] = 1
newest_data
```

**5.2 Fit the model Use1~1+BI1+Exp1**

```python
X = newest_data[['intercept','BI1','Exp1']]
y = newest_data['Use1']
fit1 = sm.OLS(y, X).fit()
fit1.summary()
```

**5.3 Intervene Exp1 = 1 and Exp = 5** 

**Then, use fitted model above to make prediction (pointwise) for Use1**

```python
temp_data = newest_data.copy()
temp_data['Exp1'] = 1
temp_data.head(3)
```

```python
parent_target1 = temp_data[['intercept','BI1','Exp1']]
df1 = pd.concat([temp_data,fit1.predict(parent_target1)],axis = 1)
df1 = df1.rename(columns={0: "Pred_use1"})
df1
```

```python
temp_data2 = newest_data.copy()
temp_data2['Exp1'] = 5
temp_data2.head(3)
```

```python
parent_target2 = temp_data2[['intercept','BI1','Exp1']]
df2 = pd.concat([temp_data2,fit1.predict(parent_target2)],axis = 1)
df2 = df2.rename(columns={0: "Pred_use1"})
df2
```

**5.4 Here, I compute the prediction in terms of confidence interval, accomplished by using bootstrapping**

**Then, I combine point wise prediction and CI prediction together**


```python
def compute_confidence_intervals1(data, val, num_bootstraps=200, alpha=0.05):
    Ql = alpha/2 #0.025
    Qu = 1 - alpha/2 #0.975
    estimates = pd.Series(data['Use1'])
    
    for i in range(num_bootstraps):
        temp_df_ind = np.random.choice(len(data),replace = True, size = len(data))
        temp_df = data.iloc[temp_df_ind]
        
        X = temp_df[['BI1','Exp1']]
        y = temp_df['Use1']
        fit1 = sm.OLS(y, X).fit()
        
        temp_data = newest_data.copy()
        temp_data['Exp1'] = val
        parent_target1 = temp_data[['BI1','Exp1']]
        pred1 = fit1.predict(parent_target1)
        estimates = pd.concat([estimates,pred1],axis = 1)
    
    estimates = estimates.sort_values(by=1, ascending=True, axis=1)
    estimates = estimates.drop(columns=['Use1'])
    estimates['ql'] = estimates.apply(lambda x: np.percentile(x, Ql*100), axis=1)
    estimates['qu'] = estimates.apply(lambda x: np.percentile(x, Qu*100), axis=1)
    return estimates[['ql','qu']]
```

```python
quantile = compute_confidence_intervals1(newest_data, val = 1, num_bootstraps=200, alpha=0.05)
res1 = pd.concat([df1,quantile],axis = 1)[['BI1','Exp1','Pred_use1','ql','qu']].drop_duplicates().sort_values(by=['BI1'])
res1
```

```python
quantile = compute_confidence_intervals1(newest_data, val = 5, num_bootstraps=200, alpha=0.05)
res2 = pd.concat([df2,quantile],axis = 1)[['BI1','Exp1','Pred_use1','ql','qu']].drop_duplicates().sort_values(by=['BI1'])
res2 
```

**5.5 Finally, plot the pointwise predicted regression with CI included**

```python
from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots()
ax.plot(res2['BI1'],res2['Pred_use1'],label = 'Exp1 = 5')
ax.plot(res1['BI1'],res1['Pred_use1'],label = 'Exp1 = 1')
ax.fill_between(res2['BI1'], res2['ql'], res2['qu'], color='b', alpha=.1)
ax.fill_between(res1['BI1'], res1['ql'], res1['qu'], color='b', alpha=.1)
ax.set_title('Linear Fit of BI1 and Expected Use1 Given Exp1 = 1 and Exp1 = 5 With 95% Confidence Interval')
ax.set_xlabel('BI1')
ax.set_ylabel('Expected Use1')
ax.legend()
```
